```python

def update_cfg(tree,key,value, if_present = ""):
    """Updates the entries for a nested dictionary"""
    if key in tree and (if_present == "" or if_present in tree) and tree[key] != value:
        tree[key] = value
        return True
    for branch in tree.values():
        if isinstance(branch, dict):
            if update_cfg(branch,key,value, if_present):
                return True
        if isinstance(branch, list): #sometimes lists of dicts
            for it in branch:
                if isinstance(it, dict): 
                    if update_cfg(it,key,value, if_present):
                        return True
    return False

def change_key(dc, key, val, if_present = ""):
    while True:
        res = update_cfg(dc, key, val, if_present)
        if not res:
            break
            
nc = 4
img_scale = (1333, 800)
dataset_type = "CocoDataset"
data_root = r"F:\source\repos\YOLOX\datasets\valentin64"

classes = tuple([str(i) for i in range(nc)])

keys_to_change_global = [("num_classes", nc, ""),
                  ("max_per_img", 750, ""),
                  ("max_per_img", 5000, "nms_pre"),
                  ("classes", classes, "")]
                  #("img_scale", img_scale, "")]
for it in keys_to_change_global:
    change_key(cfg, it[0], it[1], it[2])
    
#cfg.work_dir = job_folder

#cfg.log_config.interval = 40 # Interval to print the log
#cfg.checkpoint_config.interval = 1
#cfg.evaluation.interval = 1
#cfg.evaluation.save_best='bbox_mAP'
seed = 1234
random.seed(seed)
cfg.gpu_ids = [0]
cfg.log_config.interval = 1
cfg.data.samples_per_gpu = 4 # Batch size of a single GPU used in testing
cfg.data.workers_per_gpu = 0
#cfg.total_epochs = 40
#cfg.runner.max_epochs = 40
cfg.seed = seed
cfg.work_dir = "work_dirs/valentin"
#cfg.resume_from = r"F:\source\repos\mmdetectiontest\work_dirs\valentin\latest.pth"
cfg.load_from = r"F:\source\repos\mmdetectiontest\work_dirs\valentin\best_bbox_mAP_epoch_18.pth"
cfg.evaluation.interval = 10
cfg.evaluation.save_best='bbox_mAP'
cfg.data_root = r"F:\source\repos\YOLOX\datasets\valentin64"


change_key(cfg.data.train, "ann_file", os.path.join(data_root, "annotations/instances_train2017.json"))
change_key(cfg.data.train, "img_prefix", os.path.join(data_root, "train2017") + "/")

change_key(cfg.data.val, "ann_file", os.path.join(data_root, "annotations/instances_val2017.json"))
change_key(cfg.data.val, "img_prefix", os.path.join(data_root, "val2017") + "/")

change_key(cfg.data.test, "ann_file", os.path.join(data_root, "annotations/instances_val2017.json"))
change_key(cfg.data.test, "img_prefix", os.path.join(data_root, "val2017") + "/")

cfg.data.train.classes = classes
cfg.data.val.classes = classes
cfg.data.test.classes = classes

#cfg.data.train.dataset.ann_file = os.path.join(data_root, "annotations/instances_train2017.json")
#cfg.data.train.dataset.img_prefix = os.path.join(data_root, "train2017") + "/"

#cfg.data.val.ann_file = os.path.join(data_root, "annotations/instances_val2017.json")
#cfg.data.val.img_prefix = os.path.join(data_root, "val2017") + "/"

#cfg.data.test.ann_file = os.path.join(data_root, "annotations/instances_val2017.json")
#cfg.data.test.img_prefix = os.path.join(data_root, "val2017") + "/"



cfg.runner.max_epochs = 100 # Epochs for the runner that runs the workflow 
cfg.total_epochs = 100
cfg.lr_config = dict(
    policy='CosineAnnealing', # The policy of scheduler, also support CosineAnnealing, Cyclic, etc. Refer to details of supported LrUpdater from https://github.com/open-mmlab/mmcv/blob/master/mmcv/runner/hooks/lr_updater.py#L9.
    by_epoch=False,
    warmup='linear', # The warmup policy, also support `exp` and `constant`.
    warmup_iters=500, # The number of iterations for warmup
    warmup_ratio=0.001, # The ratio of the starting learning rate used for warmup
    min_lr=1e-07)
```
